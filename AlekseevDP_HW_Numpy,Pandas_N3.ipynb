```python
import pandas as pd
```

```python
# Задание 1
# Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
# - оценка 2 и меньше - низкий рейтинг
# - оценка 4 и меньше - средний рейтинг
# - оценка 4.5 и 5 - высокий рейтинг
# Результат классификации запишите в столбец class

df = pd.read_csv('E:\\Netology\\Numpy, Pandas\\11_ml-latest-small\\ratings.csv', encoding="utf-8")
df.head()
```

```python
def classify(row):
    """функция, которая классифицирует фильмы из материалов занятия по следующим правилам:
        - оценка 2 и меньше - низкий рейтинг
        - оценка 4 и меньше - средний рейтинг
        - оценка 4.5 и 5 - высокий рейтинг"""
    if row['rating'] <= 2:
        classification = "низкий"
    elif row['rating'] <= 4:
        classification = "средний"
    else:
        classification = "высокий"
    return classification
```

```python
# передаем строку датафрейма в функцию classify и создаем новое поле "class" с классификацией рейтинга
df['class'] = df.apply(classify, axis=1)
df.head(30)
```

```python
# Задание 2
# Используем файл keywords.csv.
# Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность 
# определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ 
# пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
# geo_data = {
# 'Центр': ['москва', 'тула', 'ярославль'],
# 'Северо-Запад': ['петербург', 'псков', 'мурманск'],
# 'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
# }
# Результат классификации запишите в отдельный столбец region.

df_keywords = pd.read_csv('E:\\Netology\\Numpy, Pandas\\11_ml-latest-small\\keywords.csv', encoding="utf-8")
df_keywords.head()
```

```python
def geo_classify(param):
    geo_data = {
        'Центр': ['москва', 'тула', 'ярославль'],
        'Северо-Запад': ['петербург', 'псков', 'мурманск'],
        'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    
    classified_region = 'undefined'
    phrase = param.replace(' ', '') #убираем все пробелы из входящей фразы param для корректного учёта склонений и вариантов: 
#     петербург - петербурга, петербурге и т.д; мурманск - мурманска, мурманской области и т.д.
    word = ''
     
    for key,value in geo_data.items(): #перебор элементов словаря
        if (value[0] in phrase) or (value[1] in phrase) or (value[2] in phrase): #поиск одного из значений словаря в обработанной фразе
            classified_region = key #если нашли значение, то региону присваиваем значение ключа и выходим из цикла
            break                    
                  
    return classified_region #возвращаем из функции значение региона
```

```python
df_keywords['region'] = df_keywords['keyword'].apply(geo_classify)
df_keywords['region'].value_counts()
```
