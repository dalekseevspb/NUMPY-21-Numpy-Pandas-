```python
# Задание 1
# Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. 
# Определите, какому фильму было выставлено больше всего оценок 5.0.
df = pd.read_csv('E:\\Netology\\Numpy, Pandas\\ml-latest-small\\ratings.csv', encoding="utf-8")
df.head()
```

```python
# создаем датафрейм с позициями с максимальным рейтингом фильма
df_filtered = df[df['rating'] == df['rating'].max()]
df_filtered.head()
```

```python
#подсчитываем количество фильмов (movieId)
result = df_filtered.value_counts('movieId')
result.head()

# Ответ: первое значение в списке - искомый результат (movieId = 318) - Побег из Шоушенка"
# 318,"Shawshank Redemption, The (1994)",Crime|Drama"     
```

```python
# Задание 2
# По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) 
# категорий 4, 12 и 21 за период с 2005 по 2010 года. 
# Не учитывайте в расчетах отрицательные значения quantity.

df_p = pd.read_csv('E:\\Netology\\Numpy, Pandas\\10._Основы_pandas\\power.csv', encoding="utf-8")
df_p.info()
```

```python
filtered_values = df_p[ ((df_p['country']=='Estonia') | (df_p['country']=='Latvia') | (df_p['country']=='Lithuania'))
                          & ((df_p['category']==4) | (df_p['category']==12) | (df_p['category']==21))
                          & ((df_p['year']>=2005) & (df_p['year']<=2010))
                          & (df_p['quantity']>0)
                         ]

print("Суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года, без учета отриц.значений quantity, равно: ", 
      filtered_values['quantity'].sum())
```
